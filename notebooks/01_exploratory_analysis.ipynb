# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# plotting style
sns.set_theme(style="whitegrid")
%matplotlib inline
```

# Load the Data

```python
data=pd.read_csv("../data/Low-alloy-steel-data.csv")
data
```

# Look up the data

```python
data.info()
data.head(20)
```

# Fixing Column names

```python
data.columns=data.columns.str.strip()
data = data.rename(columns={'Alloy code': 'alloy_id'})
data = data.rename(columns={'Temperature (°C)': 'Temperature_C'})
for col in data.columns[15:]:
    words=col.split(' ')
    words[-1]=words[-1].strip('()')
    data = data.rename(columns={col: '_'.join(words)})
data.columns
```

# Exploratory Data Analysis
Temperature vs Mechanical Properties (for same alloy)

```python
alloys=data["alloy_id"].unique()
print(alloys.shape[0])

dataByAlloyCode=data.groupby("alloy_id")

properties=['0.2%_Proof_Stress_MPa','Tensile_Strength_MPa', 'Elongation_%', 'Reduction_in_Area_%']

corr_list=[]
for alloy in alloys:
    df=dataByAlloyCode.get_group(alloy)
    corrs = df[properties + ['Temperature_C']].corr()['Temperature_C'][properties]
    corrs['alloy_id'] = alloy
    corr_list.append(corrs)

corr_df = pd.DataFrame(corr_list).set_index('alloy_id')
print(corr_df)

for i in range(0,5):
    plt.figure(figsize=(12,6))
    sns.heatmap(corr_df[19*i:19*(i+1)], annot=True, cmap='coolwarm', center=0)
    plt.title(f'Heatmap of alloys of chunk #{i+1}')
    plt.show()
```

In the **4th chunk** of alloys, we notice that for **Reduction in Area (%)**:  

- **CbC** has a correlation of **-0.15**  
- **CbA** has a correlation of **0.28**  

while most other alloys in this chunk have correlations ranging from **0.6 to 0.9**.  

It can be wrong data


```python
print(dataByAlloyCode.get_group('CbC')['Reduction_in_Area_%'].describe())
dataByAlloyCode.get_group('CbC')
```

replacing outlier with mean as it seems like a wrong entry

```python
val=dataByAlloyCode.get_group('CbC')['Reduction_in_Area_%'].mean().round(0)
idx = dataByAlloyCode.get_group('CbC').index[-1]  
data.loc[idx, 'Reduction_in_Area_%'] = val   
```

```python
dataByAlloyCode=data.groupby("alloy_id")

df=dataByAlloyCode.get_group('CbC')
corrs = df[properties + ['Temperature_C']].corr()['Temperature_C'][properties]
print(corrs)
corrs['alloy_id'] = 'CbC'
corr_list[74]=corrs
corr_df = pd.DataFrame(corr_list).set_index('alloy_id')

for i in range(3,4):
    plt.figure(figsize=(12,6))
    sns.heatmap(corr_df[19*i:19*(i+1)], annot=True, cmap='coolwarm', center=0)
    plt.title(f'Heatmap of alloys of chunk #{i+1}')
    plt.show()
```

```python
meanPropByTemp=data.groupby('Temperature_C')[properties].mean().round(2)
meanPropByTemp.plot(kind='line',figsize=(12,6))
plt.xlabel('Temperature (°C)')
plt.ylabel('Average Property Value')
plt.title('Average Mechanical Properties vs Temperature (All Alloys)')
plt.legend(title='Property')
plt.show()
```

Tensile strength and 0.2% proof stress decrease noticeably with increasing temperature — consistent with thermal softening, since higher temperatures enhance dislocation mobility, promote recovery and annihilation, and thus lower yield strength.

Elongation (%) and Reduction in Area (%) show weak dependence on temperature in the aggregated data, suggesting that ductility is influenced more by microstructural factors (e.g., grain size, phase distribution, carbide morphology) than by the temperature range tested.

# Analysis on Carbon content 

```python
fig,ax=plt.subplots(1,2,figsize=(12,6))

data[data['Temperature_C']==27].plot(kind='scatter',x='C',y='Tensile_Strength_MPa',ax=ax[0],title='C vs Tensile Strength(at 27°C)')
data[data['Temperature_C']==27].plot(kind='scatter',x='C',y='Elongation_%',ax=ax[1], title='C vs Elongation(at 27°C)')
plt.show()
```

```python
mean_props = data[data['Temperature_C']==27].groupby('C')[['Tensile_Strength_MPa','Elongation_%']].mean().reset_index()

plt.figure(figsize=(12,6))
plt.plot(mean_props['C'], mean_props['Tensile_Strength_MPa'], marker='o', label='Tensile Strength')
plt.plot(mean_props['C'], mean_props['Elongation_%'], marker='s', label='Elongation')

plt.xlabel('Carbon Content (%)')
plt.ylabel('Average Property Value')
plt.title('Average Tensile Strength and Elongation vs Carbon at 27°C')
plt.legend()
plt.grid(True)
plt.show()
```

# Key Findings
1. Correlation Heatmaps (Mechanical Properties vs Temperature by Alloy)
- Most alloys show a strong negative correlation between temperature and both tensile strength and proof stress, meaning these properties decrease as temperature increases.
- Elongation and reduction in area generally have weaker or inconsistent correlations with temperature.
- Outlier detected: For alloy CbC, reduction in area had an unexpected negative correlation, which was identified and corrected.
2. Average Mechanical Properties vs Temperature (All Alloys)
- Tensile Strength and Proof Stress decrease noticeably as temperature rises.
- Elongation and Reduction in Area remain relatively stable across temperatures, indicating these ductility measures are less sensitive to temperature changes.
3. Carbon Content Analysis at 27°C
- Tensile Strength increases with higher carbon content, showing a positive relationship.
- Elongation decreases as carbon content increases, indicating a trade-off between strength and ductility.
